```python
# Define a custom function to extract the common part
def extract_common(value):
    parts = value.split('||')
    common_part = parts[0]
    for part in parts:
        if part != common_part:
            return value  # Return the original value if any part is different
    return common_part
```

```python
data['segmentsCabinCode'] = data['segmentsCabinCode'].apply(extract_common)
```

Feature engineering and encoding:

```python
#converting flighDate into datetime datatype:
data['flightDate'] = pd.to_datetime(data['flightDate'])

#Extracting and creating new features out of the date variable
data['flightYear'] = data['flightDate'].dt.year
data['flightMonth'] = data['flightDate'].dt.month
data['flightDay'] = data['flightDate'].dt.day
```

```python
data.tail()
```

lets get departure time from the segmentsDepartureTimeRaw feature:

```python
# Function to extract the first time from the time_values column
def extract_time(time_value):
    times = time_value.split("||")
    if times:
        return times[0].split("T")[1]
    return ""

# Apply the function to the DataFrame and create a new column with the first times
data['departureTime'] = data['segmentsDepartureTimeRaw'].apply(extract_time)
```

```python
data.head()
```

```python
# Function to extract hours and minutes
def extract_hours(departure_time):
    parts = departure_time.split(':')
    if len(parts) >= 2:
        return int(parts[0])
    return None

def extract_minutes(departure_time):
    parts = departure_time.split(':')
    if len(parts) >= 3:
        minutes = parts[1].split('.')[0] 
        return int(minutes)
    return None

data['hour'] = data['departureTime'].apply(extract_hours)
data['minute'] = data['departureTime'].apply(extract_minutes)
```

Pipeline for modelling:

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, TargetEncoder
from joblib import dump, load
```

```python
#transforming numeric columns using num_transformer
num_transformer = Pipeline(
    steps=[
        ('scaler', StandardScaler())
    ]
)
```

```python
#transforming categorical columns using cat_transformer
cat_transformer = Pipeline(
    steps=[
        ('target_encoder', TargetEncoder(random_state=8))
    ]
)
```

```python
#preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num_transformer', num_transformer, num_cols),
        ('cat_transformer', cat_transformer, cat_cols),
    ]
)
```

```python
preprocessor_pipe = Pipeline(
    steps=[
        ('preprocessor', preprocessor)
    ]
)
dump(preprocessor_pipe, 'preprocessor_pipe.joblib')
```

```python
#implementing pipeline:
import xgboost as xgb
xgb_pipe = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('xgb', xgb.XGBRegressor(random_state=42))
    ]
)
```

```python
#fitting the pipeline
xgb_pipe.fit(X_train, y_train)
```

```python
#predicting y_train
y_train_preds = xgb_pipe.predict(X_train)
print('Mean Squared Error: ' , mse(y_train, y_train_preds, squared=False))
print('Mean Absolute Error: ' ,mae(y_train, y_train_preds))
```

```python
#predicting y_val
y_val_preds = xgb_pipe.predict(X_val)
print('Mean Squared Error: ',mse(y_val, y_val_preds, squared=False))
print('Mean Absolute Error: ',mae(y_val, y_val_preds)) 
```

It can be observed that the xgb regressor performed well on the unseen validation set as it did not overfit on the unseen data and gave almost the same mse and mae scores. Now lets check for the test data

Checking on the test data: 

```python
#predicting y_test
y_test_preds = xgb_pipe.predict(X_test)
print('Mean Squared Error: ',mse(y_test, y_test_preds, squared=False))
print('Mean Absolute Error: ',mae(y_test, y_test_preds))  
```

Our xgb regressor model performed great on the unseen testing data as well. No overfitting is observed
